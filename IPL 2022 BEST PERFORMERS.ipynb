```python
from bs4 import BeautifulSoup as bs
import requests
import numpy as np
```

```python
scrape_url = 'https://stats.espncricinfo.com/ci/engine/records/averages/bowling.html?id=14452;type=tournament'
r =  requests.get(scrape_url)

#Convert to a beautiful soup object
soup = bs(r.content)

#print out the html
contents = soup.prettify()
print(contents)
```

```python
most_runs = soup.find(class_ = "engineTable")
print(most_runs.prettify())
```

```python
info_rows = most_runs.find_all("tr")
for row in info_rows:
    print (row.prettify())
```

```python
import pandas as pd
```

```python
info = []
header = []
i = 0
for index,row in enumerate(info_rows):
    if index == 0:
        header= row.get_text().split('\n')[1:-1]
        print(row)
        df1 = pd.DataFrame(index = range(0,340), columns = header)
    if index != 0:
        info.append(row.get_text().split('\n')[1:-1])
        print(row)
        #names[1:]= row.find('td').get_text()
        if index % 2 == 0:
            df1.loc[i]= info[i]
            i+=2
#for i <= len(info):
    
print(header)
print(info)
print(index)
```

```python
df_bowling = df1.dropna()
```

```python
df_bowling= df_bowling.reset_index(drop=True)
```

```python
i=0
team = []
while i < len(info):
    if i % 2 != 0:
        team.append(info[i][0][1:-1])
    i+=1

df_bowling.insert(1,'Team',team)    
    
    
    
```

```python
df_bowling['Wkts'] = df_bowling['Wkts'].replace(np.nan ,0, regex = True)
df_bowling['Wkts'] = df_bowling['Wkts'].replace( '-',0, regex = True)
```

```python
df_bowling['Econ'] = df_bowling['Econ'].replace( '-',0, regex = True)
```

```python
for col in df_bowling.columns:
    if col not in ['BBI', 'Player','Team']:
        if df_bowling[col].dtype!= float:
            try:
                df_bowling[col] = df_bowling[col].astype('float32')
            except Exception as e:
                df_bowling[col] = df_bowling[col].replace( '-',0, regex = True)
                df_bowling[col] = df_bowling[col].astype('float32')   
```
